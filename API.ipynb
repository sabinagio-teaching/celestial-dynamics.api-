```python
import requests
import pandas as pd
import datetime as dt
import os
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("API_KEY")

all_rows = []

for i in range(90):   #
    date = dt.date.today() - dt.timedelta(days=i)
    date_str = date.strftime("%Y-%m-%d")

    url = f"https://api.nasa.gov/neo/rest/v1/feed"
    params = {
        "start_date": date_str,
        "end_date": date_str,
        "api_key": api_key
    }

    response = requests.get(url, params=params)
    data = response.json()

    for ast in data["near_earth_objects"][date_str]:
        row = {
            "name": ast["name"],
            "id": ast["id"],
            "hazardous": ast["is_potentially_hazardous_asteroid"],
            "close_approach_date": date_str,
            "miss_distance_km": ast["close_approach_data"][0]["miss_distance"]["kilometers"],
            "relative_velocity_km_s": ast["close_approach_data"][0]["relative_velocity"]["kilometers_per_second"],
            "diameter_min_m": ast["estimated_diameter"]["meters"]["estimated_diameter_min"],
            "diameter_max_m": ast["estimated_diameter"]["meters"]["estimated_diameter_max"]
        }
        all_rows.append(row)

df = pd.DataFrame(all_rows)
df.to_csv("neo_90days.csv", index=False)

df.head(50)
```
